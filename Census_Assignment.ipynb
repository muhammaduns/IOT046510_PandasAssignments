# Reading files through glob

```python
# reading files through glob
from glob import glob

files = glob('states*.csv') # creating array of file names
dframe = []
for afile in files:
    df = pd.read_csv(afile, index_col=None, header = 0)
    dframe.append(df)

us_census = pd.concat(dframe, axis =0 , ignore_index=True)
us_census.head(5)
```

# Droping Unnamed: 0 Column

```python
us_census.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
us_census.head(5)
```

# Spliting the GenderPop

```python
us_census[['Male','Female']] = us_census.GenderPop.str.split("_",expand=True)
```

```python
us_census.head(5)
```

# Droping column GenderPop

```python
us_census.drop(columns = ['GenderPop'], inplace=True)
```

```python
us_census.head(5)
```

# Triming / Striping $ sign from column

```python
us_census['Income'] = us_census.Income.str.strip('$')
```

```python
us_census.head(5)
```

# Removing % from columns

```python
us_census = us_census.replace('%','', regex=True)
```

```python
us_census.head(5)
```

# Removing M and F from Male and Female columns

```python
us_census = us_census.replace('F','', regex=True)
```

```python
us_census = us_census.replace('M','', regex=True)
```

```python
us_census.head(5)
```

```python
us_census.dtypes
```

# Rounding off to decimal 2 and converting to numric 

```python
us_census.loc[:,'Hispanic':'Income'] = round(us_census.loc[:,'Hispanic':'Income'].apply(pd.to_numeric),2)
```

```python
us_census.head(5)
```

```python
us_census.dtypes
```

```python
us_census['Male'] = us_census['Male'].astype(int)
```

```python
np.where(us_census.applymap(lambda x: x == ''))
```

```python
us_census.drop(columns = ['Female'], inplace=True)
```

```python
us_census['Female'] = us_census['TotalPop'] - us_census['Male']
```

```python
us_census.head(5)
```

```python
np.where(us_census.applymap(lambda x: x == ''))
```

```python
us_census.dtypes
```

```python
import matplotlib.pyplot as plt
plt.scatter(us_census['Female'],us_census['Income'])
plt.xlabel('Salary_$', fontsize=18)
plt.ylabel('Female_Population', fontsize=16)
plt.show()
```

```python
us_census.duplicated().any()
```

# Dropping duplicate values

```python
us_census.drop_duplicates(inplace= True)
```

```python
us_census.duplicated().any()
```
